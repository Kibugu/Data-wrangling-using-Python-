# 4. Filtering data using Pandas.

When you want to concentrate on several attributes of the data we can use a this function. cleaning data where the entries appear more than ones in a dataset.

```python
filtered_No=repeated_No[repeated_No>2].to_frame().reset_index() # Deletes the repeated nos
```

```python
filtered_df=df[~df.No.isin(filtered_No.No)] # Replaces the existing column with the new columns which has no repeated elements
```

The importance of a tilde "~" its the pandas boolean series is reversed and thus resulting data frame is of those that do not repeat more than twice.
incase we did not put the tilde the repeated individuals would appear twice.

When we want to add a column in the existing data frame this is achieved through the following ways using pandas.
In our above example we want to add the total premium form life insurance in kenya, then assign the market share based on that.


```python
# convert the data from object to integer and replacing the missing values with a 0.
df['2010'] = pd.to_numeric(df['2010'], errors='coerce').fillna(0)
df['2010'] = df['2010'].astype(np.int64)
df['2011'] = pd.to_numeric(df['2011'], errors='coerce').fillna(0)
df['2011'] = df['2011'].astype(np.int64)
df['2012'] = pd.to_numeric(df['2012'], errors='coerce').fillna(0)
df['2012'] = df['2012'].astype(np.int64)
df['2013'] = pd.to_numeric(df['2013'], errors='coerce').fillna(0)
df['2013'] = df['2013'].astype(np.int64)
df['2014'] = pd.to_numeric(df['2014'], errors='coerce').fillna(0)
df['2014'] = df['2014'].astype(np.int64)
df['2015'] = pd.to_numeric(df['2013'], errors='coerce').fillna(0)
df['2015'] = df['2015'].astype(np.int64)
df['2016'] = pd.to_numeric(df['2016'], errors='coerce').fillna(0)
df['2016'] = df['2016'].astype(np.int64)
df['2017'] = pd.to_numeric(df['2017'], errors='coerce').fillna(0)
df['2017'] = df['2017'].astype(np.int64)
df['2018'] = pd.to_numeric(df['2018'], errors='coerce').fillna(0)
df['2018'] = df['2018'].astype(np.int64)
```

```python
col_list= list(df)
```

```python
df['Total Premium'] = df[col_list].sum(axis=1)
```
